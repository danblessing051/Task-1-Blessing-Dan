```python
import pandas as pd
import io

df = pd.read_csv(io.StringIO(uploaded['Dataset for Data Analytics - Sheet1 (2).csv'].decode('utf-8')))

print(f"Rows: {len(df)}")
print(f"Columns: {len(df.columns)}")
print(df.head())
```

```python
print("=== COLUMN NAMES ===")
print(df.columns.tolist())


print("\n=== MISSING VALUES ===")
print(df.isnull().sum())

print("\n=== FIRST 5 ROWS ===")
print(df.head())
```

```python
# Count duplicates
duplicates = df.duplicated(subset='OrderID').sum()
print(f"Duplicate OrderIDs found: {duplicates}")

# Remove duplicates
df = df.drop_duplicates(subset='OrderID', keep =  'first')
print(f"Rows after removing duplicates: {len(df)}")
```

```python
# Check before
print("Missing CouponCode BEFORE:", df['CouponCode'].isnull().sum())

# Fill missing values
df['CouponCode'] = df['CouponCode'].fillna('NO COUPON')

# Check after
print("Missing CouponCode AFTER:", df['CouponCode'].isnull().sum())
```

```python
# Check before
print("Date format BEFORE:", df['Date'].dtype)
print(df['Date'].head(3))

# Convert to YYYY-MM-DD
df['Date']= pd.to_datetime(df['Date']).dt.strftime('%Y/%m/%d')

# Check after
print("\nDate format AFTER:")
print(df['Date'].head(3))
```

```python
# Check before
print("TotalPrice BEFORE:")
print(df['TotalPrice'].head(10))

# Round to 2 decimal places
df['TotalPrice'] = df['TotalPrice'].round(2)
df['UnitPrice'] =df['UnitPrice'].round(2)

# Check after
print("\nTotalPrice AFTER:")
print(df['TotalPrice'].head(10))
```

```python
# Find all text columns
text_cols = df.select_dtypes(include='object').columns
print(f"Text columns found: {list(text_cols)}")

# Strip whitespace fron all of them
for col in text_cols:
    df[col] = df[col].str.strip()

print("\nWhitespace stripped from all text columns!")
```

```python
# Re-applying preprocessing steps to ensure df is in the correct state

# Fill missing CouponCode values
df['CouponCode'] = df['CouponCode'].fillna('NO COUPON')
print("Missing CouponCode AFTER re-fill:", df['CouponCode'].isnull().sum())

# Convert 'Date' column to YYYY/MM/DD format
df['Date'] = pd.to_datetime(df['Date']).dt.strftime('%Y/%m/%d')
print("Date format AFTER re-conversion:")
print(df['Date'].head(3))
```

```python
print("=== FINAL VERIFICATION ===")
print(f"Total rows: {len(df)}")
print(f"Missing values anywhere: {df.isnull().sum().sum()}")
print(f"Duplicate OrderIDs: {df.duplicated('OrderID').sum()}")
print(f"Date sample: {df['Date'].head(3).tolist()}")
print(f"TotalPrice sample: {df['TotalPrice'].head(5).tolist()}")

print("\n== VERIFICATION GATE RESULT ===")
print(f"0% duplicate IDs: {'PASSED' if df.duplicated('OrderID').sum() == 0 else 'FAILED'}")
print(f"0% date errors: {'PASSED' if df['Date'].str.match(r'\d{4}/\d{2}/\d{2}').all() else 'FAILED'}")
print(f"0% missing values: {'PASSED' if df.isnull().sum().sum() == 0 else 'FAILED'}")
print(f"100% data retained: {'PASSED' if len(df) == 1200 else 'FAILED'}")
```

```python
# Save cleaned file
df.to_excel("Dataset_Cleaned.xlsx", index=False)
print("Clean file saved!")

# Download it to my computer
from google.colab import files
files.download("Dataset_Cleaned.xlsx")
```
